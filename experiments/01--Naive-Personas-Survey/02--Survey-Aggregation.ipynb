```python
import pandas as pd

import llm_moral_values as llmv
```

```python
questionnaire = llmv.Questionnaire(path="../../data/questionnaires/moral_foundations")
```

```python
dataset = llmv.pipelines.EvaluateSurvey.collate_data("./data/**/*.json")
llmv.pipelines.EvaluateSurvey.write_data_report(dataset, "./data/dataset.report.txt")
dataset.to_parquet("./data/dataset.parquet")
dataset
```

```python
llmv.pipelines.EvaluateSurvey.get_response_freq(dataset).T
```

```python
dataset_agg_dimension: pd.DataFrame = llmv.pipelines.EvaluateSurvey.aggregate_by_group(
    dataset, ["model", "persona", "dimension"]
).T
dataset_agg_dimension.to_parquet("./data/dataset.agg.dimension.parquet")
dataset_agg_dimension
```

```python
dataset_agg_question: pd.DataFrame = llmv.pipelines.EvaluateSurvey.aggregate_by_group(
    dataset, ["segment", "id", "model", "persona"]
)
dataset_agg_question.to_parquet("./data/dataset.agg.question.parquet")
dataset_agg_question
```

```python
cross_evaluation = llmv.pipelines.EvaluateSurvey.perform_cross_evaluation(
    dataset, questionnaire.get_survey("graham_et_al")
)
cross_evaluation.to_parquet("./data/dataset.cross_evaluation.parquet")
cross_evaluation
```
