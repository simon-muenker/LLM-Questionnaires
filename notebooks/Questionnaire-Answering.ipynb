```python
import glob
import itertools
import json
import pathlib
import uuid

import tqdm

import config
```

```python
ITERATIONS: int = 50

CFG = config.Config(
    data_dir=pathlib.Path("../data/"),
    report_dir=pathlib.Path("../report/"),
)
CFG
```

```python
"Total number of questionnaires filled: " + str(
    len(CFG.models) *
    len(CFG.personas) *
    ITERATIONS
)
```

```python
for model, persona, questionnaire in list(
        itertools.product(
            CFG.models.items(),
            CFG.personas.items(),
            CFG.questionnaires.items()
        )
):
    prod_path = CFG.report_dir / questionnaire[0] / persona[0] / model[0]
    prod_path.mkdir(parents=True, exist_ok=True)
    
    while len(glob.glob(f"{prod_path}/*.json")) < ITERATIONS:
        json.dump([
            dict(item, persona=persona[1].id) for item in tqdm.tqdm(
                questionnaire[1].answer_questionnaire(
                    model[1], prefix=persona[1].content
                ),
                total=len(questionnaire[1]),
                desc=f"{(model[0], persona[0], questionnaire[0])}"
            )], 
            open(prod_path / f"{uuid.uuid4()}.json", "w", encoding="utf8"),
            indent=4,
            ensure_ascii=False
        )
```
